```python
# Import required dependencies
from sentence_transformers import SentenceTransformer
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
from time import time
```

```python
# Load required model
albert_base_encoder = SentenceTransformer("albert-base-v2")
roberta_base_encoder = SentenceTransformer("roberta-base")
bert_base_encoder = SentenceTransformer('bert-base-nli-mean-tokens')
distil_roberta_encoder = SentenceTransformer('nli-distilroberta-base-v2')
```

```python
# Loading Dataset
train=pd.read_csv('drive/MyDrive/Dataset/Train_Test/train.csv')
test=pd.read_csv('drive/MyDrive/Dataset/Train_Test/test.csv')
```

```python
#creating a copy
testing=train.copy()
testing_size=30
testing=testing[0:testing_size]
```

```python
#cleaning
del testing['id']
del testing['qid1']
del testing['qid2']
del testing['is_duplicate']
```

```python
#creating corpus and queries
corpus=testing['question1']
query=testing['question2']
```

```python
# Load corpus-data
corpus_data = []
corpus=corpus.astype(str).values.flatten().tolist()
for i in range(testing_size):
  corpus_data.append(corpus[i])
```

```python
# Load query-data
queries = []
query=query.astype(str).values.flatten().tolist()
for i in range(testing_size):
  queries.append(query[i])
```

```python
# Create corpus embeddings
corpus_embed_albert_base = albert_base_encoder.encode(corpus_data)
corpus_embed_roberta_base = roberta_base_encoder.encode(corpus_data)
corpus_embed_bert_base = bert_base_encoder.encode(corpus_data)
corpus_embed_distil_roberta = distil_roberta_encoder.encode(corpus_data)
```

```python
corpus_embed_albert_base.shape, corpus_embed_roberta_base.shape , corpus_embed_bert_base.shape , corpus_embed_distil_roberta.shape
```

```python
# Function to evaluate model similarity and execution time 
def model_accuracy(encoder, query, corpus_embed):
  prev_time = time()
  query_embedding = encoder.encode(query)
  similarity_score = cosine_similarity(query_embedding.reshape(1, -1), corpus_embed)
  return (max(similarity_score[0]), time()-prev_time)
```

```python
# Evaluate albert model performance
albert_result = []
albert_performance = []
for query in queries:

  albert_sim, albert_time = model_accuracy(albert_base_encoder, query, corpus_embed_albert_base)
  albert_result.append(albert_sim)
  albert_performance.append(albert_time)
```

```python
# Evaluate robert model performance
roberta_result = []
roberta_performance = []
for query in queries:
  roberta_sim, roberta_time =  model_accuracy(roberta_base_encoder, query, corpus_embed_roberta_base)
  roberta_result.append(roberta_sim)
  roberta_performance.append(roberta_time)
```

```python
# Evaluate bert model performance
bert_result = []
bert_performance = []
for query in queries:
  bert_sim, bert_time =  model_accuracy(bert_base_encoder, query, corpus_embed_bert_base)
  bert_result.append(bert_sim)
  bert_performance.append(bert_time)
```

```python
# Evaluate distil roberta model performance
distil_roberta_result = []
distil_roberta_performance = []
for query in queries:
  distilroberta_sim, distilroberta_time =  model_accuracy(distil_roberta_encoder, query, corpus_embed_distil_roberta)
  distil_roberta_result.append(distilroberta_sim)
  distil_roberta_performance.append(distilroberta_time)
```

```python
print("Albert Result", len(albert_result), albert_result[0:3],"Albert Performance", len(albert_performance), albert_performance[0:3])
print("Roberta Result", len(roberta_result), roberta_result[0:3],"Roberta Performance", len(roberta_performance), roberta_performance[0:3])
print("Bert Result", len(bert_result), bert_result[0:3], "Bert Performance", len(bert_performance), bert_performance[0:3])
print("Distil Roberta Result", len(distil_roberta_result), distil_roberta_result[0:3], "Distil Roberta Performance",  len(distil_roberta_performance), distil_roberta_performance[0:3])
```

```python
# Create and save "albert" model performance stats
model_df = pd.DataFrame({"Model": "Albert", "Similarity Score": albert_result, "Time Taken": albert_performance})
model_df.to_csv('albert.csv')
```

```python
# Create and save "roberta" model performance stats
model_df = pd.DataFrame({"Model": "Roberta", "Similarity Score": roberta_result, "Time Taken": roberta_performance})
model_df.to_csv('roberta.csv')
```

```python
# Create and save "bert" model performance stats
model_df = pd.DataFrame({"Model": "Bert", "Similarity Score": bert_result, "Time Taken": bert_performance})
model_df.to_csv('bert.csv')
```

```python
# Create and save "distil roberta" model performance stats
model_df = pd.DataFrame({"Model": "Distil Roberta", "Similarity Score": distil_roberta_result, "Time Taken": distil_roberta_performance})
model_df.to_csv('distil_roberta.csv')
```
